## Read in all the datasets

```python
# read the creatinine measurements
df_ceratinine = pd.read_csv("./dataScienceTask/T_creatinine.csv")
# Print the shape
print("Data dimension:")
print(df_ceratinine.shape)
# Check for missing values
print("\nNumber of missing values: ")
print(df_ceratinine.isnull().sum())

# rename the value column
df_ceratinine =  df_ceratinine.rename({'value': 'vCreatinine'}, axis=1)

# Print the summary
print("\nData Summary: ")
df_ceratinine.describe()
```

The kidneys maintain the blood creatinine in a normal range. Creatinine has been found to be a fairly reliable indicator of kidney function. Normal levels of creatinine in the blood are approximately 0.6 to 1.2 milligrams (mg) per deciliter (dL) in adult males and 0.5 to 1.1 milligrams per deciliter in adult females.

**source:** https://www.medicinenet.com/creatinine_blood_test/article.htm

```python
# Visualize data 
sns.boxplot(df_ceratinine['vCreatinine'])
```

```python
# read the diastolic blood pressure measurements
df_dbp = pd.read_csv("./dataScienceTask/T_DBP.csv")
# Print the shape
print("Data dimension:")
print(df_dbp.shape)
# Check for missing values
print("\nNumber of missing values: ")
print(df_dbp.isnull().sum())

# rename the value column
df_dbp =  df_dbp.rename({'value': 'vDBP'}, axis=1)

# Print the summary
print("\nData Summary: ")
df_dbp.describe()
```

A normal diastolic blood pressure is lower than 80. But even if your diastolic number is lower than 80, you can have elevated blood pressure if the systolic reading is 120-129.

80-89 is stage 1 hypertension.

90 or more is stage 2 hypertension.

120 or more is a hypertensive crisis.

**source:** https://www.webmd.com/hypertension-high-blood-pressure/guide/diastolic-and-systolic-blood-pressure-know-your-numbers#1

The diastolic reading, or the bottom number, is the pressure in the arteries when the heart rests between beats. This is the time when the heart fills with blood and gets oxygen.

A normal systolic pressure is below 120.

A reading of 120-129 is elevated.

130-139 is stage 1 high blood pressure (also called hypertension).

140 or more is stage 2 hypertension.

180 or more is a hypertensive crisis. 

**source:** https://www.webmd.com/hypertension-high-blood-pressure/guide/diastolic-and-systolic-blood-pressure-know-your-numbers#1

## Merge the dataset

#### First all the measurements are merged on ID and time 

```python
merge1 = pd.merge(df_ceratinine, df_dbp, how='outer', on=['id', 'time'])

merge2 = pd.merge(merge1, df_gluc, how = 'outer', on = ['id', 'time'])

merge3 = pd.merge(merge2, df_hgb, how = 'outer', on = ['id', 'time'])

merge4 = pd.merge(merge3, df_ldl, how = 'outer', on = ['id', 'time'])

merge5 = pd.merge(merge4, df_sbp, how = 'outer', on = ['id', 'time'])
```

```python
## Missing value imputation
#merge5['new_vSBP'] = merge5.groupby("id").transform(lambda x: x.fillna(x.mean()))['vSBP']
```

```python
# Using time variates as one of the covariates, no need to worry about the time difference.    
```

```python
## First remove rows with all NA values 
merge5 = merge5.dropna(axis = 0, how="all")
# Check the percentage of missing values in each column
merge5.isnull().sum()/ merge5.shape[0]
# There are no missing values greater than 80% hence keep all the columns
```

### Next merge on the Patient information and Stage information

```python
df_patientInfo = pd.merge(df_demo, df_stage, how = 'left', on = ['id'])
df_patientInfo.columns
```

```python
# Plot the age distribution 
df1 = df_patientInfo.groupby(['age']).count()
df_patientInfo.iloc[:,3].hist()
```

```python
# function to plot bar chart for different variables in the patientInfo dataframe
def bar_chart(feature):
    progressed = df_patientInfo[df_patientInfo['Stage_Progress']== True][feature].value_counts()
    not_progressed = df_patientInfo[df_patientInfo['Stage_Progress']==False][feature].value_counts()
    df_bar = pd.DataFrame([progressed,not_progressed])
    df_bar.index = ['progressed','not_progressed']
    df_bar.plot(kind='bar',stacked=True, figsize=(8,5))
```

```python
# Plotting stacked bar chart for gender
bar_chart('gender')
plt.xlabel('Stage')
plt.ylabel('Number of patients')
plt.legend()
plt.title('Gender and  Stage Progress')
```

```python
# Plotting stacked bar chart for race
bar_chart('race')
plt.xlabel('Stage')
plt.ylabel('Number of patients')
plt.legend()
plt.title('Race and  Stage Progress')
```

```python
# Histogram for age for stage progress
x = [df_patientInfo.age[df_patientInfo.Stage_Progress == sp_name] for sp_name in [True, False]]
plt.hist(x, 8, density=True, histtype='bar', stacked=True)
plt.legend([True, False])
plt.title("Stage Progression by Age")
plt.xlabel('Age')
plt.ylabel('Number of patients')
```

```python
# histogram for ages.
# Each bar is an age value in the data
x1 = df_patientInfo.loc[df_patientInfo.Stage_Progress == True, 'age']
x2 = df_patientInfo.loc[df_patientInfo.Stage_Progress == False, 'age']

kwargs = dict(alpha=0.5, bins=100)
plt.figsize=(8,8)
plt.hist(x1, **kwargs, color='r', label='True')
plt.hist(x2, **kwargs, color='b', label='False')

plt.gca().set(title='Frequency Histogram of Age', ylabel='Frequency', xlabel = 'age')
plt.xlim(40,90)

plt.legend();
```

```python
# Identifying the proportion of stage progress = T/F by gender and race
print(df_patientInfo.groupby(['gender', 'Stage_Progress']).count())
df_patientInfo.groupby(['race', 'Stage_Progress']).count()
```

## Examine lab measurements by Stage Progress

```python
# Merge stages and merge5
labs = pd.merge(merge5, df_patientInfo, how = "left",  on=['id'])
```

```python
sns.boxplot(x="Stage_Progress", y="vCreatinine", hue= 'gender',
                    data=labs, palette="Set2")
# sns.swarmplot(x="Stage_Progress", y="vCreatinine",
#                     data=labs, color=".2")
```

```python
sns.boxplot(x="Stage_Progress", y="vDBP", hue= 'gender',
                    data=labs, palette="Set2")
```

```python
sns.boxplot(x="Stage_Progress", y="vGlucose", hue= 'gender',
                    data=labs, palette="Set2")
```

```python
labs.columns
```

```python
## Visualization of each of the lab measurments over time for all patients
# Initialize the figure
plt.style.use('seaborn-darkgrid')

plt.figure(figsize=(20,10))
# create a color palette
palette = plt.get_cmap('Set1')
 
# multiple line plot
num=0
for column in labs.drop(['id', 'time','race', 'gender'], axis=1):
    num+=1
 
    # Find the right spot on the plot
    plt.subplot(3,3, num)
 
    # plot every groups, but discreet
    #for v in labs.drop(['id', 'time','race', 'gender'], axis=1):
        #plt.plot(labs['age'], labs[v], marker='', color='grey', linewidth=0.6, alpha=0.3)
 
    # Plot the lineplot
    plt.plot(labs['time'], labs[column], marker='', color=palette(num), linewidth=2.4, alpha=0.9, label=column)
 
    # Same limits for everybody!
    plt.xlim(0,max(labs.time))
    plt.ylim(-2,max(labs[column]))
 
    # Not ticks everywhere
    if num in range(7) :
        plt.tick_params(labelbottom='off')
    if num not in [1,4,7] :
        plt.tick_params(labelleft='off')
 
    # Add title
    plt.title(column, loc='left', fontsize=12, fontweight=0, color=palette(num) )
 
# general title
plt.suptitle(" Lab measurment variations for Patient ID 0 stage = True", fontsize=13, fontweight=0, color='black', style='italic', y=1.02)
 

```

```python
## Lab measuremnt variation over time for patient ID == 1 ( false case)
df_sub = merge5[merge5.id == 1]
# Initialize the figure
plt.style.use('seaborn-darkgrid')

plt.figure(figsize=(20,10))
# create a color palette
palette = plt.get_cmap('Set1')
 
# multiple line plot
num=0
for column in df_sub.drop(['id', 'time'], axis=1):
    num+=1
 
    # Find the right spot on the plot
    plt.subplot(3,3, num)
 
    
    # Plot the lineplot
    plt.plot(df_sub['time'], df_sub[column], marker='', color=palette(num), linewidth=2.4, alpha=0.9, label=column)
 
    # Same limits for everybody!
    plt.xlim(0,max(df_sub.time))
    plt.ylim(-2,max(df_sub[column]))
 
    # Not ticks everywhere
    if num in range(7) :
        plt.tick_params(labelbottom='off')
    if num not in [1,4,7] :
        plt.tick_params(labelleft='off')
 
    # Add title
    plt.title(column, loc='left', fontsize=12, fontweight=0, color=palette(num))
 
# general title
plt.suptitle(" Lab measurment variations for Patient ID 1, stage = False", 
             fontsize=13, fontweight=0, color='black', style='italic', y=1.02)
 
plt.xlabel("time")
```

```python
# Plot HGB and Creatinine over 2 year time period for progress = True

for i in range(300):
    sub_data = labs[(labs.id == i) & (labs.Stage_Progress == True) ]
    plt.plot(sub_data.time, sub_data.vHGB, alpha=0.5, linewidth= 1)
    plt.title('Stage Pregress- True')
    plt.xlabel('time')
    plt.ylabel('HGB')

plt.show()
```

```python
# Plot HGB and Creatinine over 2 year time period for progress = False
for i in range(300):
    sub_data = labs[(labs.id == i) & (labs.Stage_Progress == False) ]
    plt.plot(sub_data.time, sub_data.vHGB, alpha=0.5, linewidth= 1)
    plt.title('Stage Progress- False')
    plt.xlabel('time')
    plt.ylabel('HGB')

plt.show()
```
