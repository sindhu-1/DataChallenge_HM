## Merge the dataset

#### First all the measurements are merged on ID and time 

```python
merge1 = pd.merge(df_ceratinine, df_dbp, how='outer', on=['id', 'time'])
```

```python
merge2 = pd.merge(merge1, df_gluc, how = 'outer', on = ['id', 'time'])
```

```python
merge3 = pd.merge(merge2, df_hgb[df_hgb.time <=699], how = 'outer', on = ['id', 'time'])
```

```python
merge4 = pd.merge(merge3, df_ldl, how = 'outer', on = ['id', 'time'])
```

```python
merge5 = pd.merge(merge4, df_sbp, how = 'outer', on = ['id', 'time'])
```

```python
### Merge the Medicine information with the measurements

trial = pd.merge(merge5[['id', 'time']], df_meds, how = 'left', on =['id'])
```

```python
merge5.shape
```

```python
trial = trial.join(pd.DataFrame(
   0.0, 
    index=trial.index, 
    columns= df_meds.drug.unique() + "_dose"
))
```

```python
trial.info()
```

```python
for index, row in trial.iterrows():
    if (row['start_day'] <= row['time']) | ( row['time'] <= row['end_day']):
        #
        col_name = row['drug'] + '_dose'
        val = row['daily_dosage']
        trial[col_name][index] = val
#         print(row['daily_dosage'])
#         print(row[col_name])
        
```
